# Chemical descriptors defined in RDKit

https://www.rekit.org/docs/source/rdkit.Chem.rdMolDescriptors.himl

## Descriptors related to Lipinski's rule of five

 1. CalcNumLipinskiHBA --> H-bond acceptor의 개수   ( acceptor의 개수 <=10)
 2. CalcNumLipinskiHBD --> H-bond. donor의 개수   ( donor의 개수 <=5)>)
 3. CalcCrippenDescriptors --> LogP 값 계산  ( LogP <= 5)
 4. CalcExactMolWt -->  <=500 Dalton

```python
import rdkit
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem.rdMolDescriptors import *
# rdMolDescriptors 안에 있는 모든 함수를 import
```

```python
# molecule.smi 파일에 있는 분자들의 성질 계산

from rdkit import Chem
molecules = []
for smi in open("molecule.smi").readlines():
    m = Chem.MolFromSmiles(smi)
    molecules.append(m)
```

```python
# Example) Hydrogen-bond acceptor의 개수를 count

hba_values = []
for m in molecules:
    hba = CalcNumLipinskiHBA(m) # molecule의 h-bond acceptor 개수를 계산
    hba_values.append(hba)
```

```python
print(hba_values)
```

```python
# Drawing histogram (정말로 hba_values 에서 h-bond acceptor의 개수가 10보다 작은지 알아보기 위함)
import matplotlib.pyplot as plt
```

```python
plt.hist(hba_values)
plt.xlabel("acceptors's count", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
```

```python
# If one wants to change the number of bins, use bins option

plt.hist(hba_values, bins=20)
plt.xlabel("acceptors's count", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

plt.savefig("acceptors count.pdf")
```

```python
# LogP predicted value 계산

LogP_values = []
for m in molecules:
    logp, mr = CalcCrippenDescriptors(m)  # 주의: return 값이 2개다!
    LogP_values.append(logp)
```

```python
print(LogP_values)
```

```python
plt.hist(LogP_values)
```

```python
# Histogram 범위 설정

plt.hist(LogP_values, bins=50, range=(-10, 15)) # 범위는 -10~15, 데이터를 50개로 쪼개라

plt.savefig("LogP_values_frequency")
```
