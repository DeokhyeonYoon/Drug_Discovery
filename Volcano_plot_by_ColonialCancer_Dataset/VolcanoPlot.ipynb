```python
import pandas as pd
import matplotlib.pyplot as plt
import bioinfokit
from bioinfokit import analys, visuz
```

```python
df = pd.read_csv("GEO2R tool results.csv")
```

```python
df1 = pd.DataFrame(df)
df1
```

 * adj.P.Val : 교정된 P Value
 * P.Value : raw P.Value
 * logFC(log2 fold change) : 효과 크기에 대한 추정치(예, 대조군과 비교군 사이 유전자 또는 Transcript의 발현이 얼마나 변했는지 나타냄.)

```python
bioinfokit.visuz.GeneExpression.volcano
```

```python
bioinfokit.visuz.GeneExpression.volcano(df=df1, lfc="logFC", pv="adj.P.Val", plotlegend=True, legendpos="upper right", lfc_thr=(1,2), pv_thr=(0.05, 0.05), show=True)
```

```python
bioinfokit.visuz.GeneExpression.volcano(df=df1, lfc="logFC", pv="adj.P.Val", plotlegend=True, legendpos="upper right", lfc_thr=(1,2), pv_thr=(0.05, 0.05), color=("blue", "grey", "red"), show=True)
```

```python
bioinfokit.visuz.GeneExpression.volcano(df=df1, lfc="logFC", pv="adj.P.Val", plotlegend=True, legendpos="upper right", lfc_thr=(1,2), pv_thr=(0.05, 0.05), color=("blue", "grey", "red"), geneid="Gene.Symbol", genenames=("TRPM5" ,"CHST6"), show=True)
# or geneid = "ID", genenames="11719434_a_at", "11735477_a_at", ...
```
